```python
import pandas as pd
from decimal import Decimal 
from pprint import pprint
```

```python
replacement_labels = {
    'Año': 'Ano',
    'Precio Cierre': 'Preco_fechamento',
    'Cambio %': 'Cambio%',
    'Promedio': 'Media',
    'Mínimo': 'Minimo',
    'Máximo': 'Maximo'
}

cambio_usd_brl = pd.read_html('https://www.dineroeneltiempo.com/divisas/usd-brl/historico?utm_source=chatgpt.com', encoding='UTF-8')[0]

cambio_usd_brl.rename(columns=replacement_labels, inplace=True)

cambio_usd_brl['Preco_fechamento'] = cambio_usd_brl['Preco_fechamento'].round(2)
cambio_usd_brl['Media'] = cambio_usd_brl['Media'].round(2)
cambio_usd_brl['Minimo'] = cambio_usd_brl['Minimo'].round(2)
cambio_usd_brl['Maximo'] = cambio_usd_brl['Maximo'].round(2)

cambio_usd_brl
```

```python
tables_exp = {}

for year in range(2009, 2024 + 1):
    exp_sample = pd.read_html(f'http://vitibrasil.cnpuv.embrapa.br/index.php?ano={year}&opcao=opt_06&subopcao=subopt_01', encoding='UTF-8')[3]
    exp_sample.dropna(inplace=True)
    exp_sample['Ano'] = year
    exp_sample = exp_sample.loc[exp_sample['Países'] != 'Total']
    new_exp_sample = pd.merge(exp_sample, cambio_usd_brl, on='Ano')
    tables_exp[f'exp_{year}'] = new_exp_sample
```

```python
replacement_dict ={'Alemanha, República Democrática': 'Alemanha',
                 'Cayman, Ilhas': 'Ilhas Cayman',
                 'Cocos (Keeling), Ilhas': 'Ilhas Cocos',
                 'Eslovaca, Republica': 'Eslováquia',
                 'Marshall, Ilhas': 'Ilhas Marshall',
                 'Tcheca, República': 'República Tcheca',
                 'Taiwan (FORMOSA)': 'Taiwan',
                 'Coreia, Republica Sul': 'Coréia do Sul'}


mapa_continente = {
    'Rússia': 'Europa',
    'Turquia': 'Ásia',
    'Cazaquistão': 'Ásia',
}

continentes = pd.read_html("https://paintmaps.com/pt/informacoes-do-pais/continente", encoding="UTF-8")[0]
continentes['CONTINENTE'] = (
    continentes['PAÍS']
    .map(mapa_continente)
    .fillna(continentes['CONTINENTE'])
)
```

```python
# Tratando variáveis:
densidade_vm = 0.995 

for year in range(2009, 2024 + 1):
    df = tables_exp[f'exp_{year}'].copy()
    
    # ===== Tratando variáveis =====
    df['Quantidade (Kg)'] = (
        df['Quantidade (Kg)']
        .astype(str)
        .str.strip()
        .replace('-', '0')
        .str.replace('.', '', regex=False)
    )

    df['Valor (US$)'] = (
        df['Valor (US$)']
        .astype(str)
        .str.strip()
        .replace('-', '0')
        .str.replace('.', '', regex=False)
    )

    # ===== Convertendo variáveis =====

    df['Quantidade (Kg)'] = pd.to_numeric(
        df['Quantidade (Kg)'], 
        errors='coerce'
    ).astype('Int64')


    df['Valor (US$)'] = pd.to_numeric(
        df['Valor (US$)'],
        errors='coerce'
    ).astype('Float64')

    # ===== Substituindo textos =====
    df['Países'] = df['Países'].replace(replacement_dict)

    # ===== Criando novas variáveis =====

    # litros = kg / densidade
    df['Quantidade (L)'] = (df['Quantidade (Kg)'] / densidade_vm).round(2)

    df['Valor (R$)'] = (df['Valor (US$)'] * df['Media']).round(2)
    
    df['Valor (L) US$'] = (
    (df['Valor (US$)'] / df['Quantidade (L)'])
    .where(df['Quantidade (L)'] > 0, 0)
    .round(2)
    )
    
    valor_em_reais = df['Valor (US$)'] * df['Media']
    preco_por_litro = valor_em_reais / df['Quantidade (L)']
    
    df['Valor (L) R$'] = (
        preco_por_litro
        .where(df['Quantidade (L)'] > 0, 0)
        .round(2)
    )
    

    df = df.merge(continentes,  left_on="Países", right_on="PAÍS")
    
    total_amount = df['Valor (US$)'].sum()
    df['Market Share'] = (df['Valor (US$)'] / total_amount * 100).round(2)

    df['Quantidade_Volume'] = 'Sem Volume'
    
    mask = df['Quantidade (L)'] > 0
    
    df.loc[mask, 'Quantidade_Volume'] = pd.qcut(
        df.loc[mask, 'Quantidade (L)'],
        q=4,
        labels=['Muito Baixo', 'Baixo', 'Médio', 'Alto']
    )

    # salva de volta
    tables_exp[f'exp_{year}'] = df.drop('PAÍS', axis=1)
    
```
